```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("gcore_result_all.csv")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
fig = df['comments'].plot()
fig.grid(True)
plt.xlabel('Artical No.')
plt.ylabel('# of Comments')
```

```python
fig = df['likes'].plot()
fig.grid(True)
plt.xlabel('Artical No.')
plt.ylabel('# of Likes')
```

```python
fig = df.loc[:, ['likes','comments']].plot()
fig.grid(True)
plt.xlabel('Artical No.')
plt.ylabel('Count')
```

```python
plt.grid()
plt.scatter(df['likes'], df['comments'])
plt.xlabel('# of Likes')
plt.ylabel('# of Comments')
```

```python
from sklearn import linear_model

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
X = np.asarray(df['likes'])
Y = np.asarray(df['comments'])
regr.fit(X.reshape(-1,1),Y.reshape(-1,1) )

print('Coefficients: \n', regr.coef_)
print('Intercept: \n', regr.intercept_)

fig = plt.figure()
plt.scatter(X.reshape(-1,1), Y.reshape(-1,1))
plt.plot(X.reshape(-1,1), regr.predict(X.reshape(-1,1)), color='blue',
         linewidth=3)

ax = fig.add_subplot(111)
ax.text(20, 400, r'Comments = Likes * 0.778 + 13.416', fontsize=15)

plt.grid()
plt.xlabel('# of Likes')
plt.ylabel('# of Comments')
```

```python
df['title_len'] = df['title'].str.len()
df['subtitle_len'] = df['subtitle'].str.len()
df
```

```python
from sklearn import linear_model

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
X = np.asarray(df['title_len'])
Y = np.asarray(df['likes'])
regr.fit(X.reshape(-1,1),Y.reshape(-1,1) )

print('Coefficients: {}'.format(regr.coef_[0][0]))
print('Intercept: {}\n'.format(regr.intercept_[0]))

fig = plt.figure()
plt.scatter(X.reshape(-1,1), Y.reshape(-1,1))
plt.plot(X.reshape(-1,1), regr.predict(X.reshape(-1,1)), color='blue',
         linewidth=3)

ax = fig.add_subplot(111)
ax.text(20, 300, r'Likes = Title_len * {0:.2f} + {1:.2f}'.format(regr.coef_[0][0], regr.intercept_[0]), fontsize=15)

plt.grid()
plt.xlabel('Title_len')
plt.ylabel('# of Likes')
plt.title('# of Likes Vs Title_len')
```

## article number group by years

```python
df['year'] = df['time'].str[0:4]
g1 = df.groupby(['year'])
g1.groups.keys()
```

```python
g1.count()['time'].tolist()
```

```python
g1.count()['time'].index.tolist()
```

```python
objects = g1.count()['time'].index.tolist()[::-1]
y_pos = np.arange(len(objects))
performance = g1.count()['time'].tolist()[::-1]
 
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.grid()

plt.xticks(y_pos, objects)
plt.xlabel('Year')
plt.ylabel('# of articles')
plt.title('# of Article Vs Year')
```
